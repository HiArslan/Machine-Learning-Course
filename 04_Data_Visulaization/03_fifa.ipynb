```python
import pandas as pd 
import plotly.express as px 
```

```python
df = pd.read_csv('fifa.csv')
```

```python
df.isnull().sum()
```

```python
fig = px.line(df , x='Date' , y=['ARG', 'BRA', 'ESP', 'FRA', 'GER', 'ITA'] , title="FIFA WORLD CUP GOALS RECORD (1995-2015)" ,
         template='plotly_dark')
fig.update_yaxes(title_text='Goals')
fig.update_xaxes(title_text='Date')
```

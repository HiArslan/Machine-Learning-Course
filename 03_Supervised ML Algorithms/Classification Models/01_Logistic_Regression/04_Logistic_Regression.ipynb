```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
df = pd.read_csv('Invistico_Airline.csv')
df.head(4)
```

```python
df.describe().T
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop(['Arrival Delay in Minutes'],axis=1,inplace=True)
df.drop(['Class'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df['Customer Type'].values
df['Type of Travel'].values
```

```python
df['satisfaction'] = df['satisfaction'].replace('satisfaction',1)
df['satisfaction'] = df['satisfaction'].replace('dissatisfied',0)
df['Gender'] = df['Gender'].replace('Male',1)
df['Gender'] = df['Gender'].replace('Female',0)
df['Customer Type'] = df['Customer Type'].replace('Loyal Customer',1)
df['Customer Type'] = df['Customer Type'].replace('disloyal Customer',0)
df['Type of Travel'] = df['Type of Travel'].replace('Personal Travel',1)
df['Type of Travel'] = df['Type of Travel'].replace('Business travel',0)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
# X = df.drop(['satisfaction'], axis=1)
# X
# OR
X = df.iloc[:,1:20]
X
```

```python
y = df.iloc[:,1]
y
```

```python
sc = StandardScaler()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=0)
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
print("Accuracy Score=",accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
cf = confusion_matrix(y_pred,y_test)
cf
```

# __Importing Libraries__

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### __Creating data through numpy__

```python
x=np.array([0,10])
y=np.array([0,300])
```

#### __Basic plot__

```python
plt.plot(x,y)
plt.xlabel('Freq')
plt.ylabel('Count')
```

#### __Basic plot with some marker__

```python
plt.plot(x,y,'o')
plt.xlabel('Freq')
plt.ylabel('Count')
```

#### __Creating another data through numpy__

```python
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([2,4,6,8,10,12,14,16])
```

```python
plt.plot(x,y,'o')
plt.xlabel('Freq')
plt.ylabel('Count')
```

```python
y=np.array([34,50,78,90,76])
```

```python
plt.plot(y,marker='o')
```

#### __Format and marker size__

```python
plt.plot(y,'o:g',ms=10) # ms is for marker size 
```

#### __Marker highlighting color__

```python
plt.plot(y,'o:g',ms=10,mec='r') # mec is for highlighting the markers
```

#### __Bar Graph__

```python
plt.bar([0.25,0.50,0.75,0.90],[20,30,10,40],label='Population',color='r',width=0.1)
plt.xlabel('Percent')
plt.ylabel('Freq')
plt.show()
```

#### __Histogram__

```python
age = [12,15,32,56,78,90,23,40]
bins = [0,10,20,30,100]
```

```python
plt.hist(age , bins=bins , histtype='bar' , rwidth=0.8)
```

#### __Scatter plot__

```python
weigth=np.array([1,2,3,4,5,6,7,8])
speed=np.array([6,5,4,3,2,1,0.5,0.25])
age=np.array([12,16,28,35,49,57,76,90])
```

```python
plt.scatter(weigth,speed)
```

```python
plt.plot([],[],color='g',label='weight',linewidth=5)
plt.plot([],[],color='r',label='speed',linewidth=5)
plt.stackplot(weigth,speed,age,colors=['g','y'])
```

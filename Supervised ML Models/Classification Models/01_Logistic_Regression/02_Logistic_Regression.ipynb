```python
df.drop(['glucose'],axis=1,inplace=True)
```

```python
X = df.iloc[:,[1,3]] 
y = df.iloc[: , 8]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
X_test
```

```python
plt.scatter(X_test[:,0],y_test,c=y_pred)
px.scatter(X_test[:,0] , y_test , color=y_pred )
```

```python
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
from termcolor import colored
# print(colored("Accuracy Score=" , ''),accuracy_score(y_test,y_pred))
print("\33[36m Accuracy Score=",accuracy_score(y_test,y_pred))
# print(accuracy_score(y_test,y_pred))
print(colored("Classification Report" , 'yellow').center(100))
print(classification_report(y_test,y_pred))
```

```python
cf = confusion_matrix(y_test,y_pred)
print(cf)
```
